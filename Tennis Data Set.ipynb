# A Journey of Federer from Jan 2000 to December 2016
## EDA of data set
### Import libraries

```python
import pandas as pd
import os
cwd = os.getcwd()
```

### Load tennis dataframe

```python
tennis_df = pd.read_csv(os.path.join(cwd, 'datasets\Data.csv'), encoding = "ISO-8859-1")
```

### Drop NA columns

```python
tennis_df.fillna(0, inplace=True)
```

```python
tennis_df = tennis_df.dropna(axis=1, how='any')
tennis_df
```

### How many columns are there in dataset?

```python
list(tennis_df.columns)
```

### Matches happend in between?

#### step 1: convert string column into datetime column

```python
from datetime import datetime as timeStamp
tennis_df['Date'] = tennis_df['Date'].apply(lambda str_val: timeStamp.strptime(str_val,'%d/%m/%Y').date())
```

#### step 2: find first match date

```python
timeStamp.strftime(min(list(tennis_df['Date'])), '%A(%d-%m-%Y)')
```

#### step 3: find last match date

```python
timeStamp.strftime(max(list(tennis_df['Date'])), '%A(%d-%m-%Y)')
```

### Tennis serieses held across the world

```python
list(tennis_df['Series'].unique())
```

## How many matches played in Grand Slam?

```python
tennis_df.groupby(['Series'])['ATP'].count().reset_index(name='matches')
```

### Can we play tennis in both indoor and outdoor courts?

```python
list(tennis_df['Court'].unique())
```

### Is outdoor matches are more than indoor?

```python
tennis_df.groupby(['Court'])['ATP'].count().reset_index(name='matches')
```

### Various locations matches organized

```python
list(tennis_df['Location'].unique())
```

### How many rounds in a tournament?

```python
list(tennis_df['Round'].unique())
```

### How many matches Federer R. won?

```python
tennis_df[tennis_df['Winner']=='Federer R.'].groupby('Winner')['ATP'].count().reset_index(name='matches won')
```

### How many matches Federer lost?

```python
tennis_df[tennis_df['Loser']=='Federer R.'].groupby('Loser')['ATP'].count().reset_index(name='matches lost')
```

### How many titles Federer won?

```python
tennis_df.query('Winner==\'Federer R.\' & Round==\'The Final\'').groupby('Winner')['ATP'].count().reset_index(name='Titles Won')
```

### How many finals he lost?

```python
tennis_df.query('Loser==\'Federer R.\' & Round==\'The Final\'').groupby('Loser')['ATP'].count().reset_index(name='Titles Lost')
```

### Against whom he lost most of the games?

```python
tennis_df.query('Loser==\'Federer R.\' & Round==\'The Final\'').groupby(['Winner']).agg({'ATP': 'count'}).reset_index(col_fill='No. of times')
```
